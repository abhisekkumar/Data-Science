```python
# Complete the moneyFlowIndex function below.
import numpy
import pandas as pd


def moneyFlowIndex(filename, n):
    data = pd.read_csv(filename)                                                    
    data.insert(6, "Typical Price", 0)                                          #create new column Typical Price
    data.insert(7, "Positive Money Flow", 0)                                    #create new column Positive Money Flow
    data.insert(8, "Negative Money Flow", 0)                                    #create new column Negative Money Flow
    data.insert(9, "Positive Money Flow Sum", 0)                                #create new column Positive Money Flow Sum
    data.insert(10, "Negative Money Flow Sum", 0)                               #create new column Negative Money Flow Sum
    data.insert(11, "Money Flow Index", 0)                                      #create new column Money Flow Index

    for i in range(0,len(data)):
        data.loc[i,'Typical Price'] = (data.loc[i,'High']+data.loc[i,'Low']+data.loc[i,'Close'])/3.0        #calculate Typical Price

        if i == 0:                  
            continue
        else:        
            if(data.loc[i-1,'Typical Price'] < data.loc[i,'Typical Price']):    #if the typical price is higher than the previous day price calculate positive money flow else calcualte negative money flow
                data.loc[i,'Positive Money Flow'] = data.loc[i, 'Typical Price']*data.loc[i,'Volume']               
            else:
                data.loc[i,'Negative Money Flow'] = data.loc[i, 'Typical Price']*data.loc[i,'Volume']
            
        if i < n:           
            continue
        else:
            data.loc[i,'Positive Money Flow Sum'] = sum(data.loc[i+1-n:i+1, 'Positive Money Flow'])         #calculate the PMFS for the value inclusive for n days and store i.e if n = 14,                                                                                                                               # 14+1-14:14+1 or 0 to 15 and store it in 15th index 
            data.loc[i,'Negative Money Flow Sum'] = sum(data.loc[i+1-n:i+1, 'Negative Money Flow'])
            MR = data.loc[i,'Positive Money Flow Sum']/data.loc[i,'Negative Money Flow Sum']                #Calculate Money Ration or MR
            data.loc[i,'Money Flow Index'] = (MR/(1+MR))*100                                                #Calculate Money Flow Index
    columns=['Open', 'High', 'Low', 'Close', 'Volume', 'Typical Price', 'Positive Money Flow', 'Negative Money Flow', 'Positive Money Flow Sum', 'Negative Money Flow Sum', 'Money Flow Index']
    out_str = 'money_flow_index_'+str(n)+'.csv'  
    

    
    data.to_csv(out_str, columns=columns)
    

```
