```python
data = pd.read_csv("sample.csv")
data.insert(6, "Typical Price", 0)
data.insert(7, "Positive Money Flow", 0)
data.insert(8, "Negative Money Flow", 0)
data.insert(9, "Positive Money Flow Sum", 0)
data.insert(10, "Negative Flow Sum", 0)
data.insert(11, "Money Flow Index", 0)
data.head(30)
```

```python
n = 14
for i in range(0,len(data)):
    data.loc[i,'Typical Price'] = (data.loc[i,'High']+data.loc[i,'Low']+data.loc[i,'Close'])/3.0
    #Money_Flow = Typical
    
    
    if i == 0:
        continue
    else:
        if data.loc[i,'Typical Price'] == data.loc[i-1, 'Typical Price']:
            data.drop([i])
            continue
        if(data.loc[i-1,'Typical Price'] < data.loc[i,'Typical Price']):
            data.loc[i,'Positive Money Flow'] = data.loc[i, 'Typical Price']*data.loc[i,'Volume']
        else:
            data.loc[i,'Negative Money Flow'] = data.loc[i, 'Typical Price']*data.loc[i,'Volume']
            
    if i < n:
        continue
    else:
        data.loc[i,'Positive Money Flow Sum'] = sum(data.loc[i+1-n:i+1, 'Positive Money Flow']) 
        data.loc[i,'Negative Flow Sum'] = sum(data.loc[i-n:i, 'Negative Money Flow'])
        mr = data.loc[i,'Positive Money Flow Sum']/data.loc[i,'Negative Flow Sum']
        data.loc[i,'Money Flow Index'] = (mr/(1+mr))*100
        
        
data.head(30)
```
